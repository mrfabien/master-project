This lab is only fot the variables found in SVARS, PRATE, RADVARS !

```python
import pandas as pd
import numpy as np
import xarray as xr
import os
```

```python
# import the link_era5_curnagl.txt file, where the header is the first row

servor_path = '/work/FAC/FGSE/IDYST/tbeucler/default/raw_data/ECMWF/WS_fabien/link_era5_curnagl.txt'

link = pd.read_csv(servor_path, sep="\t", header=None)

# first row is the header, name the columns

link.columns = ['name_in_curnagl','name_in_ERA5','full_name','folder']

# remove rows where column 0 = ? (missing data)

link = link[link['name_in_curnagl']!='?']

# remove first row

link = link.iloc[1:]
```

For the SVARS variables

```python
# select only the variables present in the svars folder

svars = link[link['folder']=='svars']

# extract the names of the variables in the link file

svars_variables = svars.name_in_curnagl
```

```python
# extract each of the variable from the folder /work/FAC/FGSE/IDYST/tbeucler/default/raw_data/ECMWF/ERA5/svars

# first open all the nc files in the folder vars

path = '/work/FAC/FGSE/IDYST/tbeucler/default/raw_data/ECMWF/ERA5/svars/'

# create a list of all the files in the folder

all_svars_variables = xr.open_mfdataset(path+'*.nc', combine='by_coords')
```

```python
all_svars_variables
```

```python
# tranform the selected variables into a list 

svars_variables = list(svars_variables)

svars_selected_variables = all_svars_variables[svars_variables]
svars_selected_variables

# create a dictionnary with the old names (in selected_variables) and the new names (in link.full_name)

old_names = list(svars_selected_variables.data_vars)

new_names = list(svars.full_name)

name_change = dict(zip(old_names, new_names))

# rename the variables in selected_variables

svars_selected_variables = svars_selected_variables.rename(name_change)
svars_selected_variables
```
